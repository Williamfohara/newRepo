Convert the Total_Amount into a number (not a string)

```python
df = pd.read_csv(r'/content/gdrive/MyDrive/Colab Notebooks/retail_dataset.csv')
df['total_amount'] = df['total_amount'].str.replace('$', '')
df['total_amount'] = df['total_amount'].str.replace(',', '')
df['total_amount'] = df['total_amount'].astype(float)
df.to_csv('cleaned_file.csv', index=False)
df.head(10) 
```

Remove rows with empty (NULL) values

```python
exceptions = ["address_line_1", "city", "state_province	", "postal_code"]
df = df.dropna(subset=set(df.columns) - set(exceptions))
df.head(9564)
```

Locate outliers in any of the variables. Filter them out if they are mistakes, keep them if they are exceptional cases.

```python
# Calculate the Z-score of each value in the dataframe
z = np.abs(df - df.mean()) / df.std()

# Identify rows with Z-scores greater than 3 (outliers)
outliers = df[(z > 3).any(axis=1)]

# Print the outliers
print(outliers)
```
